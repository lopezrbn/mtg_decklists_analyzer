```python
# Read and process the decklists
dfs = fun.read_decklists(deck, format)
dfs = fun.process_decklists(dfs, format)
deck_colors_list = []
n_decks_by_colors_list = []
for df in dfs:
    deck_colors_list.append(df["deck_colors"].unique()[0])
    n_decks_by_colors_list.append(df["#dl"].nunique())
n_total_decks = sum(n_decks_by_colors_list)
for i, (deck_colors, n_decks_current_colors) in enumerate(zip(deck_colors_list, n_decks_by_colors_list)):
    ratio = n_decks_current_colors / n_total_decks
    print(f"Index {i} - {deck_colors}: {n_decks_current_colors} / {n_total_decks} decks ({ratio:.2%})")
```
